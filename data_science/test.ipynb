```python
def audit_device(device):

summary = {"hostname:", "", "total_interfaces": 0, "up_count":0, "down_count":0, "down_interfaces":[]} 

for key, value in device.items(): 
    summary["hostname"] = key 
    summary["total_interfaces"] = len(value) 
   if value["status"] == "up":
     summary["up_count"] += 1 
   else:
        summary["down_count"] += 1
        summary["down_interfaces"].append(value["interfaces"])

  return summary 
```

```python
def network_summary(network):


summary= {
        "total_sites": len(network['sites']),
        "site_name": [],
        "total_devices": 0
        }

for key, value in network['sites'].items():
    summary["site_name"].append(key)
    summary["total_devices"] = len(value)

    return summary




    
```

```python
def generate_alerts(servers, threshold):

cpu_threshold = threshold.get("cpu", 0)
mem_threshold = threshold.get("memory", 0)
disk_threshold = threshold.get("disk", 0)


for key, value in servers.items():

    if value["cpu"] >= cpu_threshold:
        print(f"ALERT: {key} metric: {value['cpu']} is above threshold: {cpu_threshold}")

    if value["memory"] >= mem_threshold:
        print(f"ALERT: {key} metric: {value['memory']} is above threshold: {mem_threshold}")

    if value["disk"] >= disk_threshold:
        print(f"ALERT: {key} metric: {value['disk']} is above threshold: {disk_threshold}")
        
        
```

```python
def parse_logs(logs):
    summary = {"total": len(logs), "counts": {}, "errors": []}
    for l in logs: 
        part = l.split(" ")
        error_date = part[0]
        error_severity = part[1].replace(":", "")
        error_message = " ".join(part[2:])
        if error_severity not in summary["counts"]: 
            summary["counts"][error_severity] = 0
        else:
            summary["counts"][error_severity] += 1
        summary["errors"].append(error_message)

    return summary
        

```
